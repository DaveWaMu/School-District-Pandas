```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculations
total_schools=len(school_data_complete["school_name"].unique())
total_students=len(school_data_complete["Student ID"].unique())
total_budget=school_data["budget"].sum()
average_math=school_data_complete["math_score"].mean()
average_reading=school_data_complete["reading_score"].mean()
passing_math=school_data_complete.loc[school_data_complete["math_score"]>=70]
pct_passing_math=len(passing_math)/total_students*100
passing_reading=school_data_complete.loc[school_data_complete["reading_score"]>=70]
pct_passing_reading=len(passing_reading)/total_students*100
passing_math_reading=passing_math+passing_reading
total_passing_students=len(passing_math_reading["Student ID"].unique())
pct_overall_passing=total_passing_students/total_students*100

#New data frame
district_summary_df=pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score":[average_math],
    "Average Reading Score":[average_reading],
    "Percent Passing Math":[pct_passing_math],
    "Percent Passing Reading":[pct_passing_reading],
    "Percent Overall Passing":[pct_overall_passing]
})

#Formatting
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Average Math Score"]=round(district_summary_df["Average Math Score"],2)
district_summary_df["Average Reading Score"]=round(district_summary_df["Average Reading Score"],2)
district_summary_df["Percent Passing Math"]=round(district_summary_df["Percent Passing Math"],2)
district_summary_df["Percent Passing Reading"]=round(district_summary_df["Percent Passing Reading"],2)
district_summary_df["Percent Overall Passing"]=round(district_summary_df["Percent Overall Passing"],2)

district_summary_df
```

# Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting



```python
complete_df=pd.DataFrame(school_data_complete)

#Series
nineth_s=pd.Series=complete_df.loc[complete_df['grade']=='9th']
tenth_s=pd.Series=complete_df.loc[complete_df['grade']=='10th']
eleventh_s=pd.Series=complete_df.loc[complete_df['grade']=='11th']
twelfth_s=pd.Series=complete_df.loc[complete_df['grade']=='12th']

#Group
nineth_group=nineth_s.groupby('school_name')
tenth_group=tenth_s.groupby('school_name')
eleventh_group=eleventh_s.groupby('school_name')
twelfth_group=twelfth_s.groupby('school_name')

#Average
nineth_mean=nineth_group.mean()
tenth_mean=tenth_group.mean()
eleventh_mean=eleventh_group.mean()
twelfth_mean=twelfth_group.mean()

#Math
nineth_math=nineth_mean['math_score']
nineth=nineth_math.rename('9th')

tenth_math=tenth_mean['math_score']
tenth=tenth_math.rename('10th')

eleventh_math=eleventh_mean['math_score']
eleventh=eleventh_math.rename('11th')

twelfth_math=twelfth_mean['math_score']
twelfth=twelfth_math.rename('12th')

#Combine
math_score_df=pd.concat([nineth,tenth,eleventh,twelfth], axis=1)

#Format
math_score_df=round(math_score_df[['9th','10th','11th','12th']],2)

math_score_df
```

```python
school_summary_df
school_summary_df['Budget per Student'].sort_values()
bins=[0,597,616,635,655]
group_names=['0-597','598-616','617-635','635-655']
school_summary_df['Spending Ranges (per Student)']=pd.cut(school_summary_df['Budget per Student'],bins,labels=group_names)
school_summary_group=school_summary_df.groupby('Spending Ranges (per Student)').mean()
scores_by_spending=school_summary_group[['Average Math Score', 'Average Reading Score', 'Percent Passing Math','Percent Passing Reading']]

scores_by_spending['Overall Passing Rate']=(scores_by_spending['Percent Passing Math']+scores_by_spending['Percent Passing Reading'])/2
scores_by_spending['Overall Passing Rate']
scores_by_spending
```
